```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_tracks = pd.read_csv('../input/spotify-datasets/tracks.csv')
df_tracks.head(5)
```

```python
pd.isnull(df_tracks).sum()
```

```python
df_tracks.info()
```

```python
sorted_by_popularity = df_tracks.sort_values('popularity', ascending=True)
sorted_by_popularity.head()
```

```python
df_tracks.describe().transpose()
```

```python
most_popular = df_tracks[df_tracks['popularity'] > 90].sort_values('popularity', ascending=False)
most_popular[:5]
```

```python
df_tracks.set_index('release_date', inplace=True)
df_tracks.index = pd.to_datetime(df_tracks.index)
df_tracks.head()
```

```python
df_tracks[['artists']].iloc[2]
```

```python
df_tracks['duration'] = df_tracks['duration_ms'].apply(lambda x: round(x/1000))
df_tracks.drop('duration_ms', inplace=True, axis=1)
df_tracks[:5]
```

```python
corr_df = df_tracks.drop(['key', 'mode', 'explicit'], axis=1).corr(method='pearson')
# print(corr_df.head())
plt.figure(figsize=(14,6))
heatmap = sns.heatmap(corr_df, annot=True, fmt='.1g', vmin=-1, vmax=1, center=0, cmap='inferno', linewidth=1, linecolor='Black')
heatmap.set_title('Correlation HeatMap')
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90);
```

```python
sample_df = df_tracks.sample(int(0.005 * len(df_tracks)))
sample_df.head()
```

```python
f_size = (15, 9)
```

```python
plt.figure(figsize=f_size)
sns.regplot(data = sample_df, y = 'loudness', x = 'energy', 
            scatter_kws={"color": "c"}, line_kws={"color": "r"}).set(title='Loudness and energy');
```

```python
plt.figure(figsize=f_size)
sns.regplot(data = sample_df, y = 'popularity', x = 'acousticness', 
            scatter_kws={"color": "b"}, line_kws={"color": "r"}).set(title='Popularity and acousticness');
```

```python
plt.figure(figsize=f_size)
sns.regplot(data = sample_df, y = 'acousticness', x = 'energy', 
            scatter_kws={"color": "b"}, line_kws={"color": "r"}).set(title='Acousticness and energy');
```

```python
plt.figure(figsize=f_size)
sns.regplot(data = sample_df, y = 'valence', x = 'danceability', 
            scatter_kws={"color": "b"}, line_kws={"color": "r"}).set(title='Valence and danceability');
```

```python
df_tracks['dates'] = df_tracks.index.get_level_values('release_date')
df_tracks.dates=pd.to_datetime(df_tracks.dates)
years=df_tracks.dates.dt.year
```

```python
sns.displot(years, discrete=True, aspect=2, height=6, kind='hist').set(title='Songs per year');
```

```python
total_dr = df_tracks.duration
fig, ax = plt.subplots(figsize=(20,7))
fig = sns.barplot(x=years, y=total_dr, ax = ax, errwidth=False).set(title='Year and duration')
plt.xticks(rotation=60);
```

```python
df_genre = pd.read_csv('../input/ultimate-spotify-tracks-db/SpotifyFeatures.csv')
df_genre.head()
```

```python
plt.figure(figsize=(15,10))
plt.title('Duration of the songs by genre')
sns.color_palette('rocket', as_cmap=True)
sns.barplot(y='genre', x='duration_ms', data=df_genre)
plt.xlabel('Duration')
plt.ylabel('Genres');
```

```python
sns.set_style(style='darkgrid')
plt.figure(figsize=(10,5))
famous = df_genre.sort_values('popularity', ascending=False).head(10)
sns.barplot(y='genre', x='popularity', data=famous).set(title='Top 5 genres');
```
